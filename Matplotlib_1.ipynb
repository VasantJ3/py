```python
import matplotlib.pyplot as plt
```

```python
x = [2,5,9,4,7]
```

```python
y = [9,6,8,3,5]
```

```python
len(x)==len(y)
```

# Line Plot

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'ro') # for plotting as dots (Similar to scatter) (refer down)
plt.show()   #  by changin how you represent a line plot, you can make it into a scatter plot
```

```python
plt.plot(x,y,'r--') # for plotting as dashed lines
plt.show()
```

```python
plt.plot(x,y,'g^') # for plotting as triangles (Similar to scatter)
plt.show()   #  by changin how you represent a line plot, you can make it into a scatter plot
```

```python
plt.plot(x,y) # for plotting with labels

plt.title("Line Plot")
plt.xlabel("values of x")
plt.ylabel("values of y")
plt.show()
```

# Bar Plot

```python
plt.bar(x,y)
plt.show()
```

```python
plt.bar(x,y)

plt.title("Bar Plot")
plt.xlabel("values of x")
plt.ylabel("values of y")

plt.show()
```

# Histogram

```python
plt.hist(x)

plt.title("Histogram Plot")
plt.xlabel("values of x")
plt.ylabel("values of y")

plt.show()
```

# Scatter Plot

```python
plt.scatter(x,y)

plt.title("Scatter Plot")
plt.xlabel("values of x")
plt.ylabel("values of y")

plt.show()
```

# Multiple Plot


```python
plt.plot(x,y)

plt.show()

plt.bar(x,y)

plt.show()
```
